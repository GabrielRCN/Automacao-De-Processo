### Importar Arquivos e Bibliotecas

```python
import pandas as pd
import pathlib
import win32com.client as win32
```

```python
# Importando e Lendo a Base de Dados
email_df = pd.read_excel(r'Bases de Dados/Emails.xlsx')
lojas_df = pd.read_csv(r'Bases de Dados/Lojas.csv', encoding = 'latin1', sep = ';')
vendas_df = pd.read_excel(r'Bases de Dados/Vendas.xlsx')
```

### Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# Mesclando as Tabelas de Vendas (vendas_df) e Lojas (lojas_df) 
vendas_df = vendas_df.merge(lojas_df, on = 'ID Loja')
display(vendas_df)
```

```python
# Criando uma Tabela para Cada Loja
dicionario_lojas = {}

lista_lojas = []

for loja in vendas_df['Loja'].unique():
    lista_lojas.append(loja)
    # loc[linha, coluna]
    dicionario_lojas[loja] = vendas_df.loc[vendas_df['Loja'] == loja, :].reset_index(drop = True)
    
print('Quantidade de lojas = {}\n'.format(len(lista_lojas)))
print(lista_lojas)
```

```python
# Testando uma Loja Aleatória
display(dicionario_lojas['Shopping Center Interlagos'])
```

```python
# Definindo o Dia Indicador como o Dia Mais Recente da Base de Dados
dia_indicador = vendas_df['Data'].max()
print(dia_indicador)
```

### Salvar a planilha na pasta de backup

```python
# Caminho Para Pasta de Backup (Se a Pasta Não Estiver no Mesmo Local do Programa -> Colocar Caminho Completo)
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas')
# Lista Todos os Arquivos da Pasta de Backup
arquivos_pasta_backup = caminho_backup.iterdir()

lista_arquivos_backup = []

# Lista de Arquivos da Pasta de Backup
for arquivo in arquivos_pasta_backup:
    lista_arquivos_backup.append(arquivo.name)
print(lista_arquivos_backup)
print('\n')


# Criando as Pastas para Cada Loja
for loja in dicionario_lojas:
    # Se a Pasta Ainda Não Existir, Criar a Pasta
    if loja not in lista_arquivos_backup:
        nova_pasta = caminho_backup / loja
        nova_pasta.mkdir()
    
    # Salvando o Arquivo em Excel em Cada Pasta
    nome_arquivo = '({}-{}-{}){}.xlsx'.format(dia_indicador.year, dia_indicador.month, dia_indicador.day, loja)
    local_arquivo = caminho_backup / loja / nome_arquivo
    #print(nome_arquivo)
    dicionario_lojas[loja].to_excel(local_arquivo)
    
print('\n\nArquivos Salvos com Sucesso')
```

### Calcular os indicadores

```python
# Cálculo do Faturamento de Cada Loja
for loja in dicionario_lojas:
    faturamento_ano = dicionario_lojas[loja]['Valor Final'].sum()
    faturamento_dia = dicionario_lojas[loja].loc[dicionario_lojas[loja]['Data'] == dia_indicador, :]['Valor Final'].sum()
    print('O Faturamento Anual da Loja {} Foi de R${:,.2f}'.format(loja, faturamento_ano))
    print('O Faturamento da Loja {} no Dia {}/{}/{} Foi de R${:,.2f}'.format(loja, dia_indicador.day, dia_indicador.month, dia_indicador.year, faturamento_dia))
    print('\n')
```

```python
# Mostrar a Quantidade de Produtos Vendidos de Cada loja
for loja in dicionario_lojas:
    produto_ano = len(dicionario_lojas[loja]['Produto'].unique())
    produto_dia = len(dicionario_lojas[loja].loc[dicionario_lojas[loja]['Data'] == dia_indicador, :]['Produto'].unique())
    print('Durante o Ano, a Loja {} Vendeu {} Produtos'.format(loja, produto_ano))
    print('No Dia {}/{}/{}, a Loja {} Vendeu {} Produtos'.format(dia_indicador.day, dia_indicador.month, dia_indicador.year,loja,  produto_dia))
    print('\n')
```

```python
# Cálculo do Ticket Médio de Cada Loja
dicionario_auxiliar_lojas = dicionario_lojas.copy()

for loja in dicionario_auxiliar_lojas:
    tabela_auxiliar_ano = dicionario_auxiliar_lojas[loja].groupby('Código Venda').sum()
    ticket_medio_ano = tabela_auxiliar_ano['Valor Final'].mean()
    
    tabela_auxiliar_dia = dicionario_auxiliar_lojas[loja].loc[dicionario_lojas[loja]['Data'] == dia_indicador, :].groupby('Código Venda').sum()
    ticket_medio_dia = tabela_auxiliar_dia['Valor Final'].mean()
    
    print('O Ticket Médio Anual da Loja {} Foi de R${:,.2f}'.format(loja, ticket_medio_ano))
    print('O Ticket Médio da Loja {} no Dia {}/{}/{} Foi de RS${:,.2f}'.format(loja, dia_indicador.day, dia_indicador.month, dia_indicador.year, ticket_medio_dia))
    print('\n')
```
